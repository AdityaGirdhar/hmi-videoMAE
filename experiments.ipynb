## Visualising `train.csv`

```python
import pandas as pd
import csv

# Read the CSV file and extract labels
labels = []
with open('train.csv', 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        label = row[-1]
        labels.append(label)

# Create a pandas DataFrame with label frequencies
label_counts = pd.Series(labels).value_counts().reset_index()
label_counts.columns = ['Class', 'Frequency']
label_counts = label_counts.sort_values(by='Frequency', ascending=False)

# Print the DataFrame
label_counts
```

```python
label_counts.size
```

```python
import csv
import matplotlib.pyplot as plt

# Count occurrences of each label
label_counts = {}
for label in labels:
    if label in label_counts:
        label_counts[label] += 1
    else:
        label_counts[label] = 1

# Plot histogram
plt.figure(figsize=(10, 6))
plt.bar(label_counts.keys(), label_counts.values(), color='skyblue')
plt.xlabel('Labels')
plt.ylabel('Frequency')
plt.title('Histogram of Labels')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
from collections import Counter

# Count occurrences of each label
label_counts = Counter(labels)

# Select top N labels
N = 50 # Change N to the desired number of top labels
top_labels = dict(label_counts.most_common(N))

# Plot bar chart with top N labels
plt.figure(figsize=(10, 6))
plt.bar(top_labels.keys(), top_labels.values(), color='skyblue')
plt.xlabel('Labels')
plt.ylabel('Frequency')
plt.title(f'Top {N} Labels')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import csv
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Read the CSV file and extract labels
labels = []
with open('train.csv', 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        label = row[-1]
        labels.append(label)

# Create a string of labels separated by space
text = ' '.join(labels)

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Plot the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Labels')
plt.show()
```

## Visualising `val.csv`

```python
import pandas as pd

df = pd.read_csv('val.csv', header=None, names=['Class', 'Video_ID', 'Start_Time', 'End_Time', 'Set', 'Empty'])
df.drop(['Video_ID', 'Start_Time', 'End_Time', 'Set', 'Empty'], axis=1, inplace=True)
df.columns = ['Class']

label_counts = df.groupby('Class').size().reset_index(name='Frequency').sort_values(by='Frequency', ascending=False)
```
